### Importando Pack

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_ods_reader import read_ods
%matplotlib inline
```

```python
path = "data_time_execution.ods"
```

```python
df = read_ods(path, 1)
```

### Carregar dataSet

```python
plt.scatter(df['thread'], df['tempo (s)'])
plt.xlabel('Quantidade de Thread')
plt.ylabel('Tempo (s)')
plt.title('Thread Running Velocity')
plt.show()
```

```python
plt.plot(df['thread'], df['tempo (s)'], color='green')
plt.scatter(df['thread'], df['tempo (s)'], color = 'red')
plt.xlabel('Quantidade de Thread')
plt.ylabel('Tempo (s)')
plt.title('Thread Running Velocity')
plt.savefig("grafico_plt.png")
```

```python
df.plot(figsize = (16, 6))
plt.xlabel('Quantidade de Thread')
plt.ylabel('Tempo (s)')
plt.title('Conjunto de dados')
plt.savefig("grafico_df.jpg")
```
